## 1. My lane as an ML task (type)

Task Type: Scoring and Ranking (via Classification). While the ultimate goal is to generate a ranked queue, the core ML task is a binary classification problem. The model will output a probability score (between 0 and 1) representing the likelihood that a page needs attention. We then sort the entire dataset by this probability score in descending order to create our ranked action queue.

## 2. Target or proxy

The Proxy Target: We need a mathematical label to train against. Since we don't have a magic flag for "definitely needs a refresh," we must use a proxy.

* Starter Proxy: A binary label indicating historical decline, such as is_declining_label = (trend_direction == "down").

* Advanced Target (Future Window): A more robust target would predict a future state (e.g., using features from the prior 90 days to predict a significant traffic drop over the next 30 days).

## 3. Success metric

Success Metric: Precision@K (specifically Precision@50).
Standard accuracy is misleading here because the classes are imbalanced (most pages are fine). Furthermore, human reviewers have limited capacity. We do not care about finding every single declining page in the database (Recall); we care that the top 50 pages the model recommends are actually worth the reviewer's time. If Precision@50 is high, the tool is operationally successful.

## 5. Why ML beats a fixed rule here

A fixed, hardcoded rule (e.g., impressions > 500 AND ctr < 0.5) is rigid. It treats a 5-year-old informational blog post exactly the same as a 2-month-old product page.

Machine Learning beats a fixed rule here because it can weigh multiple interacting variables simultaneously (age, position tier, engagement rate, content type, and search volume) to find non-linear, subtle patterns. It adapts to the actual distribution of the data rather than relying on human guesswork to set arbitrary cutoffs.
