## 2. The question: decision, action, cost of a wrong call

* The decision: Which visible pages under-capture clicks or show signs of decline, and should therefore be reviewed first for a content refresh or expansion?

* The action: A human reviewer will use this ranked queue to allocate their limited time, inspecting the top-K recommended pages first to decide if a structural or content update is warranted.

* The cost of a wrong call: False positives mean wasted human reviewer time on pages that are actually fine (or dropping due to unavoidable seasonality). False negatives mean bleeding traffic on pages that could have been easily fixed.

## 3. Quick look at the data (2-3 real numbers)

"Identifying that a specific percentage of these 30,000+ pages have high 90-day impressions but disproportionately low clicks proves that a human cannot manually sift through this inventory. An ML ranking system is necessary to surface the top 50 most urgent cases."

```python
import pandas as pd

# Load the starter dataset directly from your public GitHub repo
url = 'https://raw.githubusercontent.com/NasorHidar/fly-rank-ml-1/refs/heads/main/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(url)

# 1. Total volume of pages in this starter slice
total_pages = len(df)

# 2. Find pages with high impressions but low clicks (potential CTR opportunity)
high_imp_low_click = df[(df['impressions_90d'] >= 500) & (df['clicks_90d'] < 10)]
opportunity_count = len(high_imp_low_click)

print(f"Total pages in starter dataset: {total_pages:,}")
print(f"Pages with high impressions (>=500) but low clicks (<10): {opportunity_count:,}")
print(f"Percentage of inventory needing potential review: {(opportunity_count/total_pages)*100:.1f}%")
```

## 4. Careful words: what I can and can't claim

* What I can claim: I can claim that this model identifies historical, observable patterns associated with traffic decline or underperformance, and accurately ranks these pages against a baseline rule to support decision-making.

* What I CANNOT claim: I cannot claim causal proof—updating these pages will not guarantee a traffic recovery, as this is observational data, not an experiment. I also cannot claim to be predicting Google's algorithm.
